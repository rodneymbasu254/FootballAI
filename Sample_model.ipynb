```python
data = pd.read_csv('newData.csv') 
data.head()
```

```python
from sklearn import linear_model
```

```python
X = data[['age', 'weight_kgs', 'strength', 'long_shots', 'interceptions', 'positioning', 'vision', 'penalties', 'composure', 'marking', 'standing_tackle', 'sliding_tackle']]
y = data['overall_rating']
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
print(regr.coef_)
```

```python
predictedRating = regr.predict([[26, 70, 56, 56, 87, 65, 65, 89, 45, 87, 67, 65]])
print("Predicted player rating is: ", predictedRating)
```
